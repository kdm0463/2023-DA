```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# plt.figure(figsize=(7, 4))
plt.figure(figsize=(5, 3))

x = np.linspace(-2 * np.pi, 2 * np.pi, 200) # 0에서 2*pi까지 경계선을 포함해 200개ㅐ의 수로 등분(199등분)
plt.plot(x, np.sin(x)*10, ':r')
plt.plot(x, x**2)
plt.plot(x, 2 * x**2 - 3*x)

# 전체 그림 범위 지정
plt.xlim(-10, 10)
plt.ylim(-30, 120)

plt.grid(True)
plt.title('다양한 그래프')
plt.show()
```

## boxplot()

```python
import matplotlib.pyplot as plt
import random

result = []
for i in range(10):
    result.append(random.randint(1, 1000))
    
print(sorted(result))

d = plt.boxplot(result, whis=0.5)
plt.show()
```

```python
d['whiskers'][0]
```

```python
import matplotlib.pyplot as plt
import random

result1 = []
result2 = []
for i in range(13):
    result1.append(random.randint(1, 500))
    result2.append(random.randint(501, 1000))

plt.boxplot(result1)
plt.boxplot(result2)
plt.show()
```

## subplot 그리기

```python
import matplotlib.pyplot as plt

# 서브플롯 생성
plt.subplot(2, 1, 1)  # 2행 1열의 서브플롯 중 첫 번째
plt.subplot(2, 1, 2)  # 2행 1열의 서브플롯 중 두 번째

plt.show()
```

```python
import matplotlib.pyplot as plt

# 서브플롯 생성
plt.subplot(3, 2, 1)  # 2행 1열의 서브플롯 중 첫 번째
plt.subplot(3, 2, 2)  # 2행 1열의 서브플롯 중 두 번째
# plt.subplot(3, 2, 3)  # 2행 1열의 서브플롯 중 두 번째
plt.subplot(3, 2, 4)  # 2행 1열의 서브플롯 중 두 번째
plt.subplot(3, 3, 5)  # 2행 1열의 서브플롯 중 두 번째
# plt.subplot(3, 2, 5)  # 2행 1열의 서브플롯 중 두 번째
plt.subplot(3, 2, 6)  # 2행 1열의 서브플롯 중 두 번째

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ax1 = plt.subplot(2, 1, 1)
plt.plot(np.random.randint(1, 10, 10), 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

#ax2 = plt.subplot(2, 1, 2)
ax2 = plt.subplot(2, 1, 2)
plt.plot(np.arange(10), np.arange(10)*2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)  # 2행 2열의 서브플롯

axes[0, 0].plot([1, 2, 3, 4])
axes[0, 0].set_title("Subplot 1")

axes[0, 1].plot([4, 3, 2, 1])
axes[0, 1].set_title("Subplot 2")

axes[1, 0].hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4], bins=5)
axes[1, 0].set_title("Subplot 3")

axes[1, 1].scatter([1, 2, 3, 4], [4, 3, 2, 1])
axes[1, 1].set_title("Subplot 4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)  # 2행 2열의 서브플롯

axes[0][0].plot([1, 2, 3, 4])
axes[0][0].set_title("Subplot 1")

axes[0][1].plot([4, 3, 2, 1])
axes[0][1].set_title("Subplot 2")

axes[1, 0].hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4], bins=5)
axes[1, 0].set_title("Subplot 3")

axes[1, 1].scatter([1, 3, 2, 4], [4, 3, 2, 1])
axes[1, 1].set_title("Subplot 4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```
