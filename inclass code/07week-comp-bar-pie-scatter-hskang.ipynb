```python
cubes = {i: pow(i, 3) for i in range(1, 11) if i % 2 == 0}
print(cubes)
```

```python
words = ['apple', 'banana', 'cherry', 'avocado']
first_dict = {word: word[0].upper() for word in words}
print(first_dict)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
double_dict1 = {k: v*2 for k, v in dict1.items()}
print(double_dict1)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
reverse_dict1 = {v: k for k, v in dict1.items()}
print(reverse_dict1)
```

```python
import matplotlib.pyplot as plt

# Runtime Configuration Parameters, 줄여서 rcParams
# plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rc('axes', unicode_minus = False)
# plt.rcParams['axes.unicode_minus'] = False

data = [10, 20, 30, 40]
name = ['A형', 'B형', 'AB형', 'O형']

plt.pie(data, labels = name)   # 비율 및 범례 추가
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

data = [15, 25, 35, 45]
name = ['A형', 'B형', 'AB형', 'O형']

plt.pie(data, labels = name, autopct='%.1f%%')   # 비율 및 범례 추가
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')

data = [15, 25, 35, 45]
name = ['A형', 'B형', 'AB형', 'O형']
color = ['violet', 'skyblue', 'limegreen', 'wheat']

plt.axis('equal')
plt.pie(data, labels = name, autopct = '%.1f%%', colors=color, explode=(0, 0, 0.1, 0))
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100) : 
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.scatter(x, y, s = size)
plt.grid(True)
plt.show()
```
