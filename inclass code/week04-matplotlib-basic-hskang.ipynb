```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True) # 10을 포함해서 8개 수로 등분, 전체 7등분
x2 = np.linspace(0, 10, N, endpoint=False) #  10을 빼고 8개 수로 등분, 전체 8등분
plt.plot(x1, y + 0.5, 'o', label='endpoint=True')
plt.plot(x2, y, '.', label='endpoint=False')

plt.legend()
plt.ylim([-0.5, 1]) # y 축 범위 지정
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True) # 10을 포함해서 8개 수로 등분, 전체 7등분
x2 = np.linspace(0, 10, N, endpoint=False) #  10을 빼고 8개 수로 등분, 전체 8등분
plt.plot(x1, y + 0.5, '-o', label='endpoint=True')
plt.plot(x2, y, ':.', label='endpoint=False')

plt.legend()
plt.ylim([-0.5, 1]) # y 축 범위 지정
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 20) # -10에서 10까지 (20-1)등분
y = np.zeros(20)

plt.plot(x, y, 'r*')
print(x)
```
