```python
import pandas as pd

index = pd.date_range('1/1/2024', periods=8)
df = pd.DataFrame(np.random.rand(8, 3), index=index, columns= ['A', 'B', 'C'])
df
```

```python
df['D'] = df['A'] / df['B']  # A열의 값을 B열의 값으로 나눈 값을 D열에 저장
df
```

```python
import numpy as np

df['E'] = np.sum(df, axis=1)  #행 우선 계산 값을 E 열에 저장
df
```

```python
df['F'] = np.ones(8)
df.head(3)
```

```python
df1 = df.drop(['E', 'F'], axis='columns')
df1.head()
```

```python
del df1['D']
df1.head()
```

```python
df1.columns = ['a', 'b', 'c']
df1
```

```python
df1.rename(columns={'a': '2020', 'b': '2021'}, inplace=True)
df1.tail()
```

```python
df1 = df1.sub(df1['2020'], axis='index')
df1.tail()
```

## 판다스 데이터프레임에서 부분 추출

```python
import pandas as pd
import numpy as np

data = np.random.randint(1, 11, (3, 4))
df = pd.DataFrame(data=data, index=pd.date_range('2024-01-01', periods=3),
             columns=list('ABCD'))
df
```

```python
df.loc['2024-01-01'] # 행 참조
```

```python
df.loc['2024-01-01':'2024-01-03'] # 행 참조
```

```python
df.iloc[0] # 인덱스로 행 참조
```

```python
df.iloc[:2] # 슬라이스로 여러 행 참조
```

```python
df['A'] # 열 참조
```

```python
df[['A', 'C']] # 여러 열 참조
```

```python
df.iloc[2, :] # index==2 인 행 모두 참조
```

```python
df.iloc[:, 1] # 칼럼 index==1 인 열 모두 참조
```

```python
df.iloc[:2, :2] # 행과 열 참조
```

```python
df.loc[:'2024-01-03', ['A', 'C']] # 행과 열 참조
```
