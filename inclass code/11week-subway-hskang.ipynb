```python
import csv
f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

max_station = ['', '', '', '']
print(max_station)
max_station = ['' for _ in range(4)]
print(max_station)
max_people = [0 for _ in range(4)]
print(max_people)

# index: 4, 5, 6, 7
label = ['유임승차','유임하차','무임승차','무임하차']

for row in data :
    for i in range(4, 8) :
        row[i] = int(row[i].replace(',', ''))
        if row[i] > max_people[i-4] :
            max_people[i-4] = row[i]
            max_station[i-4] = row[3] + ' ' + row[1]

f.close()

for i in range(4) :
    print(label[i] + ' : ' + max_station[i] + ':', max_people[i], '명')
```

```python
lst = ['34', '1999', '2786', '1000000']
result = map(int, lst)
print(result)
list(result)
```

```python
# map에서 내장함수 활용
my_map = map(pow, [1, 2, 3], [2, 2, 3])
list(my_map)
```

```python
def myreplace(s):
    return s.replace(',', '')

lst = ['34,000', '1,999', '2,786', '1,000,000']
result = map(myreplace, lst)
list(result)
```

```python
lst = ['34,000', '1,999', '2,786', '1,000,000']
result = map(lambda s: s.replace(',', ''), lst)
list(result)
```

```python
lst = ['34,000', '1,999', '2,786', '1,000,000']
result = map(int, map(lambda s: s.replace(',', ''), lst))
list(result)
```

```python
# map에서 인자인 시퀀스의 짝이 맞지 않아도 오류 발생하지 않고, 적은 수에 맞게 결과 나옴
list(map(lambda x, y: x+y, [1, 5, 11], [2, 8, 16, 30]))
```

```python
# 모든 지하철역의 시간대별 승하차 인원 추이 시각화

import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)
next(data)

s_in = [0] * 24
s_out = [0] * 24

for row in data :
    for i in range(4, 52):
        row[i] = row[i].replace(',', '')

    row[4:] = map(int, row[4:])
    for i in range(24) :
        s_in[i] += row[4 + i * 2]
        s_out[i] += row[5 + i * 2]

f.close()

plt.rc('font', family='Malgun Gothic')

# plt.figure(figsize=(10, 6))
# plt.style.use('ggplot')
# plt.style.use('default')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.grid(True)
timelst = map(lambda x: x%24, range(4,28))
plt.xticks(range(24), timelst)
plt.show()

print(sum(s_in))
print(sum(s_out))
```
