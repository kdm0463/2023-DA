```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

mean = []
high = []
low = []

for row in data :
    if row[-1] != '' and row[-2] != '' :
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25' :
            mean.append(float(row[-3]))
            low.append(float(row[-2]))
            high.append(float(row[-1]))

plt.figure(figsize = (10,2))
plt.plot(mean, 'k')
plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')

plt.xlabel('Day (December 25)')
plt.ylabel('The highest temperature')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

mean = []
high = []
low = []

for row in data :
    if row[-1] != '' and row[-2] != '' :
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25' :
            mean.append(float(row[-3]))
            low.append(float(row[-2]))
            high.append(float(row[-1]))

plt.figure(figsize = (10,2))
plt.plot(mean, 'k')
plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')

plt.title('크리스마스의 기온 변화 그래프')
plt.xlabel('Day (December 25)')
plt.ylabel('The highest temperature')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

temp_day = []
temp_mean = []
year = []

for row in data :
    if int(row[0].split('-')[0]) > 1907:
        if row[2] != '':
            temp_day.append(float(row[2]))
            if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '31' :
                year.append(int(row[0].split('-')[0]))
                temp_mean.append(sum(temp_day) / len(temp_day))
                temp_day = []

f.close()

plt.figure(figsize = (10,4))
plt.plot(year, temp_mean, color='k', marker='d')

plt.title('평균 기온 변화 그래프')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.grid()
# plt.grid(axis = 'x', color = 'grey', linestyle = ':', linewidth = .5)
# plt.grid(color = 'grey', linestyle = ':', linewidth = .1)
# plt.xticks(range(year[0], year[-1], 10))
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.figure(dpi = 150) # 해상도를 높이려면 기본 값인 100 이상으로
plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31) : 
    day.append([])

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day, showfliers=False)
plt.show()
```
